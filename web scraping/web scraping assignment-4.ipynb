```python
dr1=webdriver.Chrome(r'F:\Python\chromedriver.exe')
u1='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
dr1.get(u1)
dr1.maximize_window()
```

```python
rank=[]
name=[]
artist=[]
date=[]
view=[]
rk=dr1.find_elements_by_xpath('//div[@id="mw-content-text"]//table[3]//td[1]')
nm=dr1.find_elements_by_xpath('//div[@id="mw-content-text"]//table[3]//td[2]')
art=dr1.find_elements_by_xpath('//div[@id="mw-content-text"]//table[3]//td[3]')
vw=dr1.find_elements_by_xpath('//div[@id="mw-content-text"]//table[3]//td[4]')
dt=dr1.find_elements_by_xpath('//div[@id="mw-content-text"]//table[3]//td[5]')
for i in rk:
    if i.text is None :
        rank.append("--") 
    else:
        rank.append(i.text)
for i in nm:
    if i.text is None:
        name.append('--')
    else:
        name.append(i.text.replace("[]",''))
for i in vw:
    if i.text is None:
        view.append('--')
    else:
        view.append(i.text)
for i in art:
    if i.text is None:
        artist.append('--')
    else:
        artist.append(i.text)
for i in dt:
    if i.text is None:
        date.append('--')
    else:
        date.append(i.text)
youtube_video=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'Views':view,'Upload Date':date})
youtube_video
```
